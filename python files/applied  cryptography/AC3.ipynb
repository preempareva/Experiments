```python
def encrypt(text, k1, k2):
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                encrypted_char = chr((k1 * (ord(char) - 65) + k2) % 26 + 65)
            else:
                encrypted_char = chr((k1 * (ord(char) - 97) + k2) % 26 + 97)
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text

def decrypt(text, k1, k2):
    decrypted_text = ""
    mod_inv = mod_inverse(k1, 26)
    for char in text:
        if char.isalpha():
            if char.isupper():
                decrypted_char = chr(((ord(char) - 65 - k2) * mod_inv) % 26 + 65)
            else:
                decrypted_char = chr(((ord(char) - 97 - k2) * mod_inv) % 26 + 97)
            decrypted_text += decrypted_char
        else:
            decrypted_text += char
    return decrypted_text

def mod_inverse(k1, m):
    for x in range(1, m):
        if (k1 * x) % m == 1:
            return x
    return -1

def main():
    while True:
        print("Affine Cipher")
        print("1. Encrypt")
        print("2. Decrypt")
        print("3. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            text = input("Enter the text to encrypt: ")
            k1 = int(input("Enter the value of 'k1': "))
            if k1 % 2 == 0:
                print("Warning: 'k1' cannot be even. Exiting the program.")
                break
            k2 = int(input("Enter the value of 'k2': "))
            encrypted_text = encrypt(text, k1, k2)
            print("Encrypted Text:", encrypted_text)
        elif choice == 2:
            text = input("Enter the text to decrypt: ")
            k1 = int(input("Enter the value of 'k1': "))
            if k1 % 2 == 0:
                print("Warning: 'k1' cannot be even. Exiting the program.")
                break
            k2 = int(input("Enter the value of 'k2': "))
            decrypted_text = decrypt(text, k1, k2)
            print("Decrypted Text:", decrypted_text)
        elif choice == 3:
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
