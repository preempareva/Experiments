```python
def encrypt(plain_text, col):
    cipher_text = ""
    column = [""] * col
    i = 0
    while i < len(plain_text):
        for j in range(col):
            if i < len(plain_text):
                column[j] = column[j] + plain_text[i]
                i = i + 1

    cipher_text = ''.join(column)
    return cipher_text


def decrypt(cypher_text, col):
    plain_text = ""
    num=0
    rem=len(cypher_text)%col
    if(rem>0):
        num=(len(cypher_text)//col)+1
    else:
        num=(len(cypher_text)//col)


    column = [""] * num

    i = 0
    while i < len(cypher_text):
        if(rem>0):
            for j in range(num):
                if i < len(cypher_text):
                    column[j] = column[j] + cypher_text[i]
                    i = i + 1
            rem=rem-1
        else:
            for j in range(num-1):
                if i < len(cypher_text):
                    column[j] = column[j] + cypher_text[i]
                    i = i + 1

    plain_text = ''.join(column)
    return plain_text
while True:
    print("Menu")
    print("1. Encrypt")
    print("2. Decrypt")
    print("3. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
            plaintext = input("Plain text: ")
            num_columns = int(input("Enter number of columns: "))
            ciphertext = encrypt(plaintext, num_columns)
            print("Encrypted text:", ciphertext)
    elif choice == 2:
            ciphertext = input("Cipher text: ")
            num_columns = int(input("Enter number of columns: "))
            plaintext = decrypt(ciphertext, num_columns)
            print("Decrypted text:", plaintext)
    elif choice == 3:
            print("Exiting program.")
            break
    else:
            print("Invalid choice. Please enter a valid option.")
```
