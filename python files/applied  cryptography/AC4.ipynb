```python
import numpy as np
import matplotlib.pyplot as plt
population_size =  1000
sample_size =  100
num_samples =  10000
np.random.seed(42)
population = np.random.exponential(scale=1.0, size=population_size)
def draw_samples_and_compute_means(population, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        mean = np.mean(sample)
        sample_means.append(mean)
    return sample_means
sample_means = draw_samples_and_compute_means(population, sample_size, num_samples)
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Sampling Distribution of Sample Means')
plt.grid(True)
mu = np.mean(population)
sigma = np.std(population) / np.sqrt(sample_size)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax,  100)
p =  1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```

```python
# prompt: Verify Central Limit Theorem for Sampling distribution for Sample mean for randomly generated normal population using Python.

import random
import matplotlib.pyplot as plt
import statistics

# Generate a random normal population with mean 50 and standard deviation 10.
population = [random.normalvariate(50, 10) for _ in range(1000)]

# Set the sample size and number of samples.
sample_size = 30
num_samples = 1000

# Initialize an empty list to store the sample means.
sample_means = []

# Repeat the sampling process for the specified number of samples.
for _ in range(num_samples):
  # Draw a random sample from the population.
  sample = random.sample(population, sample_size)

  # Calculate the sample mean.
  sample_mean = statistics.mean(sample)

  # Add the sample mean to the list.
  sample_means.append(sample_mean)

# Plot the histogram of the sample means.
plt.hist(sample_means, bins=50)
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Mean for Randomly Generated Normal Population")
plt.show()
```

```python
import random
import statistics
import matplotlib.pyplot as plt
population_proportion = 0.395
population_size = 1000
population = [1 if random.random() < population_proportion else 0 for _ in range(population_size)]
sample_size = 30
num_samples = 1000
sample_proportions = []

# Repeat the sampling process for the specified number of times
for _ in range(num_samples):
  # Draw a random sample from the population
  sample = random.sample(population, sample_size)

  # Calculate the proportion of smokers in the sample
  sample_proportion = sum(sample) / len(sample)

  # Add the sample proportion to the list
  sample_proportions.append(sample_proportion)

# Plot the distribution of the sample proportions
plt.hist(sample_proportions, bins=50)

# Add labels and title
plt.xlabel("Sample Proportion")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Proportion for Smoking")

# Show the plot
plt.show()

# Verify Central Limit Theorem
# The distribution of the sample proportions should resemble a normal distribution, even though the population is not normally distributed.
```

```python
import numpy as np
import matplotlib.pyplot as plt
population_proportion =  0.395
population_size =  1000
sample_size =  50
num_samples =  10000
np.random.seed(42)
population = np.random.binomial(1, p=population_proportion, size=population_size)
def draw_samples_and_compute_proportions(population, sample_size, num_samples):
    sample_proportions = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        proportion = np.sum(sample) / sample_size
        sample_proportions.append(proportion)
    return sample_proportions
sample_proportions = draw_samples_and_compute_proportions(population, sample_size, num_samples)
plt.hist(sample_proportions, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Sample Proportion')
plt.ylabel('Density')
plt.title('Sampling Distribution of Sample Proportions')
plt.grid(True)
mu = np.mean(population)
sigma = np.std(population) / np.sqrt(sample_size)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax,  100)
p =  1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
plt.plot(x, p, 'k', linewidth=2)
plt.show()
print('The resulting plot should show that the distribution of sample proportions')
print('approaches a normal distribution, verifying the Central Limit Theorem')
```
