```python
def encrypt(plain_text, shift):
    cipher_text = ""
    for char in plain_text:
        if char.isalpha():
            ascii = ord('a')
        if char.islower() else ord('A')
            cipher_text += chr((ord(char) - ascii + shift) % 26 + ascii)
        else:
            cipher_text += char
    return cipher_text

def decrypt(cipher_text, shift):
    return encrypt(cipher_text, -shift)
while True:
    print("Menu")
    print("1. Encrypt")
    print("2. Decrypt")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        plain_text = input("Enter plain text: ")
        shift = int(input("Enter shift value: "))
        print("Cipher text: ", encrypt(plain_text, shift))

    elif choice == 2:
        cipher_text = input("Enter cipher text: ")
        shift = int(input("Enter shift value: "))
        print("Plain text: ", decrypt(cipher_text, shift))

    elif choice == 3:
        break
    else:
        print("Invalid choice! Please enter a valid option.")
```
